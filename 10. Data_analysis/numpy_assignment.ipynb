### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.

```python
array = np.random.randint(1, 11, size=(3,3))
print(array)

# calculating the determinant of matrix 
det = np.linalg.det(array) 
  
print("\nDeterminant of given 3X3 matrix:") 
print(int(det)) 


# calculating the inverse of a matrix
print("\n Inverse of given 3X3 matrix:") 
inverse = np.linalg.inv(array)
print(inverse)


# calculating the eigen values of a matrix
print("\n Eigen values of given 3X3 matrix:") 
eigen_values = np.linalg.eigvals(array)
print(eigen_values)
```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
import numpy.ma as ma

# creating array
array = np.random.randint(1, 30, size=(4, 4))
print("Original array: \n\n", array)

masked_arr = ma.masked_greater(array,10)
print("\nMasked array: \n \n", masked_arr)

# computing the sum of unmasked elements
sum_unmasked = masked_arr.sum()
print(f" \nSum of unmasked array is: {sum_unmasked}")
```

2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
import numpy.ma as ma

# creating array
array = np.random.randint(1, 30, size=(3,3))
print("Original array: \n\n", array)

# syntax => masked_arr = ma.masked_array(array, mask=mask)

masked_arr = ma.masked_array(array,mask=np.eye(3, dtype = bool))
print("\nMasked array: \n \n", masked_arr)

'''the dtype=bool is used to create a boolean mask where True represents masked elements, 
and False represents unmasked elements
'''


# computing the mean of unmasked elements
mean_unmasked = masked_arr.mean()
print(f" \nmean of unmasked array is: {mean_unmasked}")
```
